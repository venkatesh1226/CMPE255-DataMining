```python
import pandas as pd
from pycaret.classification import *
```

```python
data = pd.read_csv('/content/train.csv')
```

```python
clf1 = setup(data, target='genre')
```

```python
best_model = compare_models()
```

```python
from scipy.sparse import random
import gradio as gr
import random
def predict_genre(id,movie,synopsis):
    # Assuming 'data' is a dictionary with input features

    dic = {'id':[id],'movie_name': [movie], 'synopsis': [synopsis]}
    input_data = pd.DataFrame.from_dict(dic)
    result = predict_model(best_model, data=input_data)
    prediction = result['prediction_label'][0]  # Extract the predicted label
    return prediction
iface = gr.Interface(
    fn=predict_genre,
    inputs=[
        gr.inputs.Textbox(label="id"),
        gr.inputs.Textbox(label="Movie Name"),
        gr.inputs.Textbox(label="Synopsis")
    ],
     outputs=gr.outputs.Textbox(label="Predicted Genre"),
)
iface.launch(share=True, debug=True)
```
