### Step 3: Visualize Audio Waveforms

Visualizing waveforms will help us get a sense of the audio data's structure and characteristics.​

```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import random

def plot_waveform(filename, title):
    """Plot the waveform of the given audio file."""
    with wave.open(filename, 'r') as wf:
        signal = wf.readframes(-1)
        signal = np.frombuffer(signal, dtype=np.int16)
        time = np.linspace(0, len(signal) / wf.getframerate(), num=len(signal))

        plt.figure(figsize=(10, 3))
        plt.title(title)
        plt.plot(time, signal)
        plt.xlabel('Time (seconds)')
        plt.ylabel('Amplitude')
        plt.grid(True)
        plt.show()

# Randomly select a few audio files from each category for visualization
num_samples = 3
sample_cat_files = random.sample(cat_files, num_samples)
sample_dog_files = random.sample(dog_files, num_samples)

# Plot waveforms
for cat_file in sample_cat_files:
    plot_waveform(os.path.join(cats_dogs_path, cat_file), f"Waveform for {cat_file}")

for dog_file in sample_dog_files:
    plot_waveform(os.path.join(cats_dogs_path, dog_file), f"Waveform for {dog_file}")
```

### Step 4: Data Preprocessing
Given that the audio files have different lengths, one standard approach is to convert them into spectrograms, which are 2D representations of the audio signal. This makes it easier to handle variable-length audio and can provide better features for machine learning models. Spectrograms capture both frequency and time information.

Convert Audio to Spectrograms: This will transform each audio file into a 2D representation that captures frequency and time information.

Standardize Spectrogram Dimensions: Given that audio files have different durations, their corresponding spectrograms might have different widths. We'll resize them to a common dimension for consistency.

### Step 4.1: Compute Spectrograms Using wave and numpy​
We'll use the librosa library to convert each audio file into a spectrogram. Let's start by generating spectrograms for a few sample audio files to visualize the results.

```python
from scipy.signal import stft

def plot_spectrogram_alternative(filename, title):
    """Plot the spectrogram of the given audio file using numpy and wave."""
    with wave.open(filename, 'r') as wf:
        signal = wf.readframes(-1)
        signal = np.frombuffer(signal, dtype=np.int16)

        # Compute short-time Fourier transform
        f, t, Zxx = stft(signal, fs=wf.getframerate(), nperseg=1024, noverlap=512)
        magnitude = np.abs(Zxx)
        magnitude_db = 10 * np.log10(magnitude)

        plt.figure(figsize=(10, 4))
        plt.pcolormesh(t, f, magnitude_db, shading='gouraud')
        plt.colorbar(label='Intensity [dB]')
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')
        plt.title(title)
        plt.tight_layout()
        plt.show()

# Plot spectrograms for the same sample audio files using the alternative method
for cat_file in sample_cat_files:
    plot_spectrogram_alternative(os.path.join(cats_dogs_path, cat_file), f"Spectrogram for {cat_file}")

for dog_file in sample_dog_files:
    plot_spectrogram_alternative(os.path.join(cats_dogs_path, dog_file), f"Spectrogram for {dog_file}")
```

We have successfully computed and visualized the spectrograms for some sample audio files using the alternative method. The spectrograms provide a 2D representation of the frequency content of the audio signals over time.

### Step 4.2: Standardize Spectrogram Dimensions
Given that the audio files have different durations, their corresponding spectrograms might have different widths (time dimension). To standardize them, we'll choose a fixed width and either:

Pad shorter spectrograms with zeros to match the desired width.
Trim longer spectrograms to the desired width.
We'll also convert these spectrograms into a format suitable for machine learning models (e.g., flatten them into 1D arrays or resize them to standard 2D image sizes).

 we will choose a default value based on the median width of the spectrograms in the dataset.​

Compute the Widths of All Spectrograms: This will give us a sense of the distribution of widths.

Determine the Default Width: We'll use the median width as our default.

Standardize Spectrogram Dimensions: Pad or trim spectrograms to the default width.

Step 4.2.1: Compute the Widths of All Spectrograms
First, we'll compute the widths (time dimension) of all the spectrograms in the dataset to determine the median width.

```python
def compute_spectrogram_width(filename):
    """Compute the width (time dimension) of the spectrogram for the given audio file."""
    with wave.open(filename, 'r') as wf:
        signal = wf.readframes(-1)
        signal = np.frombuffer(signal, dtype=np.int16)

        # Compute short-time Fourier transform
        _, t, _ = stft(signal, fs=wf.getframerate(), nperseg=1024, noverlap=512)

        return len(t)

# Compute spectrogram widths for all audio files
all_files = cat_files + dog_files
widths = [compute_spectrogram_width(os.path.join(cats_dogs_path, file)) for file in all_files]

# Determine the median width
median_width = int(np.median(widths))
median_width
```

### Step 4.2.2: Standardize Spectrogram Dimensions
Now, we'll standardize the spectrograms by ensuring they all have a width of 186. For spectrograms with a smaller width, we'll pad them with zeros. For those with a larger width, we'll trim them to fit the desired size.

We'll also convert these spectrograms into a format suitable for machine learning models by resizing them to standard 2D image sizes (128x128 in this case).

Let's begin the standardization process.​

```python
from scipy.ndimage import zoom

def standardize_spectrogram(filename, target_width=186, image_size=(128, 128)):
    """Standardize the spectrogram dimensions and convert to an image format."""
    with wave.open(filename, 'r') as wf:
        signal = wf.readframes(-1)
        signal = np.frombuffer(signal, dtype=np.int16)

        # Compute short-time Fourier transform
        f, t, Zxx = stft(signal, fs=wf.getframerate(), nperseg=1024, noverlap=512)
        magnitude = np.abs(Zxx)

        # Pad or trim to match the target width
        current_width = magnitude.shape[1]
        if current_width < target_width:
            padding = target_width - current_width
            magnitude = np.pad(magnitude, ((0, 0), (0, padding)), mode='constant')
        else:
            magnitude = magnitude[:, :target_width]

        # Resize to image format
        resized_magnitude = zoom(magnitude, (image_size[0] / magnitude.shape[0], image_size[1] / magnitude.shape[1]))

        return resized_magnitude

# Test the function on a sample file
sample_file = os.path.join(cats_dogs_path, sample_cat_files[0])
sample_spectrogram = standardize_spectrogram(sample_file)

# Display the standardized spectrogram
plt.figure(figsize=(6, 6))
plt.imshow(sample_spectrogram, origin='lower', aspect='auto', cmap='inferno')
plt.colorbar(label='Intensity')
plt.title("Standardized Spectrogram")
plt.show()
```

The displayed image is a standardized spectrogram with dimensions 128x128, derived from one of the sample cat audio files. This format is suitable for machine learning models, especially convolutional neural networks (CNNs) used for image classification.

### Step 4.3: Create a Dataset of Standardized Spectrograms
Next, we'll convert all the audio files into standardized spectrograms and create a dataset. Each entry in the dataset will consist of the spectrogram image and its corresponding label (cat or dog).

Let's proceed with this step.​

```python
# Convert all audio files into standardized spectrograms and create a dataset
X = []  # List to store the standardized spectrograms
y = []  # List to store the labels (0 for cat, 1 for dog)

# Process cat files
for file in cat_files:
    spectrogram = standardize_spectrogram(os.path.join(cats_dogs_path, file))
    X.append(spectrogram)
    y.append(0)

# Process dog files
for file in dog_files:
    spectrogram = standardize_spectrogram(os.path.join(cats_dogs_path, file))
    X.append(spectrogram)
    y.append(1)

# Convert lists to numpy arrays
X = np.array(X)
y = np.array(y)

X.shape, y.shape
```

### Step 6: Clustering and Anomaly Detection
Before training a machine learning model, it's essential to identify and handle potential outliers or anomalies. One approach to do this is through clustering:

Clustering: Using algorithms like KMeans, we can cluster the data into groups. Ideally, for our binary classification problem, we'd expect two primary clusters (cats and dogs). However, any small, distant clusters might indicate anomalies or outliers.

Anomaly Detection: After clustering, we can identify potential anomalies by looking for data points that are distant from their cluster centers.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Flatten the data for clustering
X_train_flat = X_train.reshape((X_train.shape[0], -1))

# Apply KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42)
cluster_labels = kmeans.fit_predict(X_train_flat)

# Compute silhouette score to assess clustering quality
silhouette_avg = silhouette_score(X_train_flat, cluster_labels)

silhouette_avg
```

The silhouette score for our clustering is approximately
0.701
0.701.

The silhouette score ranges from -1 to 1, where a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters. A score close to 1 suggests strong clustering.

Given our binary classification problem (cats vs. dogs), the high silhouette score is encouraging. It suggests that the two classes are relatively distinct in the spectrogram feature space.

Step 6.1: Anomaly Detection

To detect anomalies, we'll compute the distance of each data point from its cluster center. Data points that are farther from the center than a given threshold might be considered anomalies.

Let's identify potential anomalies based on their distance from the cluster centers.​

```python
# Compute distances of each sample to its cluster center
distances = np.linalg.norm(X_train_flat - kmeans.cluster_centers_[cluster_labels], axis=1)

# Determine a threshold for anomalies (e.g., 95th percentile of distances)
threshold = np.percentile(distances, 95)

# Identify anomalies
anomalies = np.where(distances > threshold)[0]

# Number of anomalies detected
num_anomalies = len(anomalies)
num_anomalies
```

### Handling Anomalies: Imputation

For each anomalous spectrogram, we'll find the 'k' nearest non-anomalous spectrograms.
We'll then average these 'k' nearest neighbors to generate an imputed spectrogram to replace the anomalous one.

Let's set k to 5 for this exercise. We'll use the Euclidean distance as our distance metric.

Let's implement the Nearest Neighbors imputation.



```python
from sklearn.neighbors import NearestNeighbors

# Number of neighbors for imputation
k = 5

# Initialize Nearest Neighbors model
nbrs = NearestNeighbors(n_neighbors=k+1, algorithm='ball_tree').fit(X_train_flat)

# For each anomaly, find k nearest non-anomalous spectrograms
for anomaly_index in anomalies:
    distances, indices = nbrs.kneighbors([X_train_flat[anomaly_index]])

    # Exclude the first nearest neighbor because it will be the anomaly itself
    neighbor_indices = indices[0][1:]

    # Compute the mean of the k neighbors
    imputed_data = np.mean(X_train_flat[neighbor_indices], axis=0)

    # Replace the anomaly with the imputed data
    X_train_flat[anomaly_index] = imputed_data

# Check if imputation was successful for one of the anomalies
anomaly_example = anomalies[0]
original_data = X_train[anomaly_example]
imputed_data_reshaped = X_train_flat[anomaly_example].reshape(X_train.shape[1], X_train.shape[2])

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(original_data, origin='lower', aspect='auto', cmap='inferno')
plt.title("Original Anomalous Spectrogram")
plt.colorbar(label='Intensity')

plt.subplot(1, 2, 2)
plt.imshow(imputed_data_reshaped, origin='lower', aspect='auto', cmap='inferno')
plt.title("Imputed Spectrogram")
plt.colorbar(label='Intensity')

plt.tight_layout()
plt.show()
```

On the left, we have the original anomalous spectrogram, and on the right is the imputed spectrogram, which was derived by averaging the 5 nearest neighbors.

The imputed spectrogram seems to be a smoothed version, which is expected as it's an average of multiple spectrograms.

### Step 8.1: Simple Feedforward Neural Network
We'll first design and train a simple feedforward neural network on the training data. This will serve as our baseline model.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

# Design the feedforward neural network
model_ffnn = Sequential([
    Flatten(input_shape=(X_train.shape[1], X_train.shape[2])),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compile the model
model_ffnn.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Set up early stopping to prevent overfitting
early_stopping = EarlyStopping(patience=10, restore_best_weights=True)

# Train the model
history_ffnn = model_ffnn.fit(X_train, y_train, epochs=100, batch_size=32,
                              validation_split=0.2, callbacks=[early_stopping], verbose=0)

# Evaluate the model on the training and testing sets
train_acc_ffnn = model_ffnn.evaluate(X_train, y_train, verbose=0)[1]
test_acc_ffnn = model_ffnn.evaluate(X_test, y_test, verbose=0)[1]

train_acc_ffnn, test_acc_ffnn
```

To potentially improve performance and robustness, we can create an ensemble by combining the predictions of the three models. One simple approach is to use majority voting: for each sample, the ensemble prediction is the class that is predicted by at least two of the three models.
